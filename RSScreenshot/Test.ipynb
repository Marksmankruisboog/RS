```
from PIL import Image
from pytesseract import image_to_string
from google.colab import drive
import re
import os
from datetime import datetime


drive.mount('/content/gdrive')
ItemList = open("/content/gdrive/My Drive/Colab Notebooks/buylimitslong2",'r').read().split('\n')
for i in range(len(ItemList)):
    ItemList[i] = ItemList[i].split('\t')
    ItemList[i][0] = ItemList[i][0][0:-1]

```

```
def GiveName(TestName):
    List = TestName.split()
    Length = len(List)
    Max = Length
    for i in range(Length):
        TestWord = ""
        for j in range(Max):
            TestWord += List[j]
            if(j != Max-1):
                TestWord += " "
        Max -= 1
        #print(TestWord)
        for j in range(len(ItemList)):
            if(ItemList[j][0] == TestWord):
                return TestWord
            #print(TestWord)
    return "NotFound"

def ParseImage(Path):
  Tijd = os.path.getmtime(Path)
  Tijd = datetime.fromtimestamp(Tijd + 7200).strftime('%Y-%m-%d %H:%M:%S')
  ReturnList = []
  ImageStart = Image.open(Path)
  gray = ImageStart.convert('L')
  Text =  (image_to_string(gray))
  Text = Text.split('\n')
  for i in range(len(Text)):
    Word = GiveName(Text[i])
    if (Word != "NotFound"):
      Amount = re.findall('\d+',Text[i-1])
      if (len(Amount) == 0 or Text[i-1].find('for') > -1):
        Amount = 1
      else:
        Amount = Amount[0]
      Price = re.findall('\d+',Text[i+1])[0]
      ReturnList.append([Word,Amount,Price])
  return ReturnList,Tijd
  
```

```
Lijst,Tijd = ParseImage('/content/gdrive/My Drive/Colab Notebooks/Test.JPG')
print(Lijst)
print(Tijd)
```
